```python
df = df.drop(index=df.index[0])
df.head()
```

```python
df.isnull ().sum ()
```

```python
# nombre verdadero de cada variable
print(repr(df.columns.tolist()))
```

```python
#renombrar las columnas
df.rename(columns={
    "Nombre completo del líder o representante legal:": "representante_legal",
    "Nombre del Emprendimiento\\Negocio\\Empresa (si aun no lo tiene definido, escriba \"Sin definir\")": "nombre_emprendimiento",
    "Ubicación del Emprendimiento": "ubicacion_emprendimiento",
    "Tipo": "tipo_emprendimiento",
    "Canales": "canales_comercializacion"
}, inplace=True)
```

```python
# Valores distintos de cada columna
df.nunique()
```

```python
# Visualizar valores internos de cada columna
for column in df.select_dtypes(include='object').columns:
    print(f"Column: {column}")
    print(df[column].unique())
    print("\n" + "-" * 30 + "\n")
```

```python
# Quitar espacios
df['representante_legal'] = df['representante_legal'].str.strip()
df['representante_legal'] = df['representante_legal'].str.replace(r'\s+', ' ', regex=True)
```

```python
# colocar los valores en minuscula
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
#camniar los valores de la columna por una vercion normalizada
df['ubicacion_emprendimiento'] = "tunja"
```

```python
# valores atipicos de "tipo_emprendimiento"
valores_atipicos_tipo = df[(df['tipo_emprendimiento'] == '45') | (df['tipo_emprendimiento'] == '05 bochica')]
display(valores_atipicos_tipo)
```

```python
df['tipo_emprendimiento'] = df['tipo_emprendimiento'].replace(['45', '05 bochica'], 'alojamiento y servicios de comida')
```

```python
# normalizar los valores

# diccionario de los valores a normalizar
servicios_comida = [
    "restaurante", "restaurante desayuno y almuerzo", "restaurante almuerzo tipo buffet",
    "restaurante desayuno y almuerzo", "restaurante ejecutivo",
    "restaurante desayuno almuerzo y comida", "servicios de restaurante",
    "servicio de restaurante", "servicio de restaurante con especialidad en sopas tradicionales y carne a la llanera",
    "restaurante almuerzo", "desayuno almuerzo comida", "desayuno almuerzo",
    "desayuno almuerzo y comida", "venta de empanada", "venta de comida",
    "venta de comida empleados y visitantes del hospital santiago de tunja",
    "somos un restaurante de comida casera", "gastronomía", "broster", "comida broaster",
    "almuerzo y sena", "almuerzo", "almuerzos", "platos ejecutivos",
    "platos especiales y eventos", "comida rapida", "alimentos", "Servicio y variedad en el plato ejecutivo"

]

alojamiento = [
    "hospedaje", "hotel 24 horas", "único hotel spa que está en la ciudad",
    "hotel 24/7", "hotel", "recibe hospedaje", "servicio de hospedaje"
]

# Imputación según categoría
df.loc[df['canales_comercializacion'].isin(servicios_comida), 'canales_comercializacion'] = "servicio_comida"
df.loc[df['canales_comercializacion'].isin(alojamiento), 'canales_comercializacion'] = "alojamiento"
```

```python
df['canales_comercializacion'] = df['canales_comercializacion'].replace('nan', 'sin_especificar')
```

```python
unique_values = df['canales_comercializacion'].unique()
print(f"Valores únicos en 'canales_comercializacion': {unique_values}")
print(f"Cantidad de valores únicos: {len(unique_values)}")
```

```python
# eliminar la columna tipo emprendimiento es irelevante
df = df.drop(columns=['tipo_emprendimiento'])
```

```python
import plotly.express as px

# Contar la frecuencia de cada canal de comercialización
canal_counts = df['canales_comercializacion'].value_counts().reset_index()
canal_counts.columns = ['Canal de Comercialización', 'Cantidad']

# Crear el gráfico de torta
fig = px.pie(canal_counts,
             values='Cantidad',
             names='Canal de Comercialización',
             title='Distribución de Canales de Comercialización',
             hole=0.3) # Agrega un agujero para un donut chart si lo prefieres

fig.show()
```

```python
#datos no validos
total_rows = len(df)

# Filtrar las filas que no son 'servicio_comida' ni 'alojamiento'
otros_canales = df[~df['canales_comercializacion'].isin(['servicio_comida', 'alojamiento'])]

# Contar las filas que cumplen con esta condición
count_otros_canales = len(otros_canales)

# Calcular la proporción
proportion_otros_canales = (count_otros_canales / total_rows) * 100

print(f"El número de emprendimientos con canales distintos a 'servicio_comida' y 'alojamiento' es: {count_otros_canales}")
print(f"La proporción de estos valores es: {proportion_otros_canales:.2f}%")
```

```python
# proporcion de datos validos
proportion_servicio_alojamiento = 100 - proportion_otros_canales
print(f"{proportion_servicio_alojamiento:.2f}%")
```

```python
# moda categorica
mode_canales_comercializacion = df['canales_comercializacion'].mode()
print(f"La moda de la columna 'canales_comercializacion' es: {mode_canales_comercializacion.iloc[0]}")
mode_count = df['canales_comercializacion'].value_counts().loc['servicio_comida']
print(f"La moda 'servicio_comida' aparece {mode_count} veces en la columna 'canales_comercializacion'.")
```
